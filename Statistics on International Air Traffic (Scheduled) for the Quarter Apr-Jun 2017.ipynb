```python
import numpy as np
d = np.array(data_f['PASSENGERS FROM INDIA'].values)
```

```python
np.random.seed(41)
sample_means = []
for i in range(1000):
  mean = np.random.choice(d,size=35).mean()
  sample_means.append(mean)
```

```python
sns.distplot(pd.DataFrame(data=sample_means,columns=["means"]))
```

```python
from scipy.stats import norm

#compute the z_critical value
z_crit = norm.ppf(q=0.95) #For 90% confidence the q-value is 0.95

#error interval
err = z_crit * (np.std(sample_means)/np.sqrt(500))
```

```python
class Format():
  BOLDs = '\033[1m'
  BOLDe = '\033[0m'
```

```python
print("{}Z-Critical\t:{}{:.3f}".format(Format.BOLDs,Format.BOLDe,z_crit))
print("{}Error\t\t:{}{:.3f}".format(Format.BOLDs,Format.BOLDe,err))
print("{}Lower Limit\t:{}{:.3f}".format(Format.BOLDs,Format.BOLDe,np.mean(sample_means)-err))
print("{}Upper Limit\t:{}{:.3f}".format(Format.BOLDs,Format.BOLDe,np.mean(sample_means)+err))
```

```python
distribution(data_f['FREIGHT FROM INDIA'],color='g')
```

```python
from scipy.stats import ttest_1samp
sample_size = 15
freightFromInd = np.random.choice(np.array(data_f['FREIGHT FROM INDIA'].values),sample_size)
stat,p = ttest_1samp(freightFromInd,data_f['FREIGHT FROM INDIA'].mean())
if p < 0.05:
  print("Reject null hypothesis")
else:
  print("Accept Null hypothesis")
```

```python
data_f['FREIGHT FROM INDIA'].mean(),np.mean(freightFromInd)
```
